Inventory Managment System

```python
import json , datetime
date=datetime.datetime.now()
date=date.strftime('%c')
again=1
total=0
selldict={}
with open('product.json') as json_file:
    product_details = json.load(json_file)
    print(product_details)

    while again==1:
       for i in product_details:
         print('id=',i)
         print('product name = ',product_details[i]['product_name'], 'product Price = ',product_details[i]['price'], 'product Quantity = ',product_details[i]['quantity'])
       productid=input("Enter Product id")
       x=product_details[productid]
       print(x)
       pquantity=int(input("Enter number of quantity"))
       if(pquantity>x['quantity']):
         print("Out Of Stock")
       else:
           x['quantity']=x['quantity']-pquantity
           product_details[productid]=x
           pprice= (x['price']-x['price']*(x['discount']/100))*pquantity
           print(pprice)
           total=total +pprice
           y=x
           y['quantity']=pquantity
           selldict[productid]=y
       again =int(input("Enter 1 for buy more else enter 0)"))
       print(total)
       sell=json.dumps(selldict)
       js=json.dumps(product_details)
fs=open('sell.json','a')
fs.write(sell)
fs.close()       
fd=open('product.json','w')
fd.write(js)
fd.close()
print(selldict)
```
